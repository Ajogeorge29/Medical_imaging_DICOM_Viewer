```python
import pydicom
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
import os

def view_dicom(file_path):
    # Read the DICOM file
    ds = pydicom.dcmread(file_path)

    # Extract the image data
    image_data = ds.pixel_array

    # Display the image using matplotlib
    plt.imshow(image_data, cmap=plt.cm.gray)
    plt.title('DICOM Image')
    plt.axis('off')
    plt.show()

def convert_dicom(file_path, output_format='png'):
    # Read the DICOM file
    ds = pydicom.dcmread(file_path)

    # Extract the image data
    image_data = ds.pixel_array

    # Normalize the image data to the range 0-255
    image_data = (np.maximum(image_data, 0) / image_data.max()) * 255.0
    image_data = np.uint8(image_data)

    # Convert to a PIL image
    image = Image.fromarray(image_data)

    # Determine the output file path and format
    base_name = os.path.splitext(file_path)[0]
    output_file = f"{base_name}.{output_format}"

    # Save the image in the desired format
    image.save(output_file)
    print(f"Image saved as {output_file}")

if __name__ == "__main__":
    # Example usage
    dicom_file = r"D:\2024\Medical_Imaging_Projects\0000.dcm"  # Raw string literal to avoid escape sequence issues
    
    # View the DICOM file
    view_dicom(dicom_file)
    
    # Convert the DICOM file to PNG
    convert_dicom(dicom_file, output_format='png')
    
    # Convert the DICOM file to JPEG
    convert_dicom(dicom_file, output_format='jpeg')
```
